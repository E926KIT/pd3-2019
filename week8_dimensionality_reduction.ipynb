```
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```
X.shape, y.shape
```

```
X[:5, :]
```

```
y[:5], y[-5:]
```

```
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

# X = pca.fit_transform(X)
```

```
X.shape
```

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for label in (0, 1, 2):
  x = X[y==label, :]
  ax.scatter(x[:, 0], x[:, 1], x[:, 2])
```

```
import tensorflow as tf
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```
x_train.shape, y_train.shape
```

```
i = 200
plt.imshow(x_train[i, :, :])
print(y_train[i])
```

```
x_train_flat = x_train.reshape((60000, 28*28))
```

```
x_train_flat.shape
```

```
from sklearn.model_selection import train_test_split

x_train_flat, _, y_train, _ = train_test_split(x_train_flat, y_train, test_size=0.95)
x_train_flat.shape, y_train.shape
```

```
from sklearn.manifold import TSNE

x_train_embedded = TSNE(n_components=2).fit_transform(x_train_flat)
```

```
fig, ax = plt.subplots()
for label in range(10):
  x = x_train_embedded[y_train==label, :]
  ax.scatter(x[:, 0], x[:, 1], label=str(label))
ax.legend()
```
