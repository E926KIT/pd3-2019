# 回帰問題

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)
x
```

```python
y = np.sqrt(x) * 10 + np.random.randn(100) * 5
plt.scatter(x, y)
```

```python
X = x[:, np.newaxis]
print(x.shape)
print(X.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
X2 = np.hstack((X, X**2))
X2[:5, :]
```

```python
model.fit(X2, y)
y_pred = model.predict(X2)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
X10 = np.hstack([X**i for i in range(1, 11)])
X10[:5, :]
```

```python
model.fit(X10, y)
y_pred = model.predict(X10)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
model.coef_
```

```python
X100 = np.hstack([X**i for i in range(1, 101)])
model.fit(X100, y)
y_pred = model.predict(X100)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
model.coef_
```

```python
plt.hist(model.coef_)
```

## L2 正則化 (Ridge Regression)
penalty = $\alpha \sum \theta_i^2$

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=1.0)
model.fit(X100, y)
y_pred = model.predict(X100)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
plt.hist(model.coef_)
```

```python
model = Ridge(alpha=1000.0)
model.fit(X100, y)
y_pred = model.predict(X100)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
plt.hist(model.coef_)
```

## L1 正則化 (Lasso Regression)
penalty = $\alpha \sum |\theta_i|$

```python
from sklearn.linear_model import Lasso

model = Lasso(alpha=1.0)
model.fit(X100, y)
y_pred = model.predict(X100)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

```python
plt.hist(model.coef_)
```

```python
model = Lasso(alpha=1000.0)
model.fit(X100, y)
y_pred = model.predict(X100)
plt.scatter(x, y)
plt.plot(x, y_pred, 'r-')
```

# 分類問題
## ロジスティック回帰

```python
from sklearn.datasets import load_iris

X, y  = load_iris(return_X_y=True)
print(X.shape)
print(y.shape)
```

```python
X[:5, :]
```

```python
y
```

```python
y[y == 0] = 2
y
```

```python
plt.scatter(X[:, 1], X[:, 2], c=y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X[:, 1:3], y)
y_pred = model.predict(X[:, 1:3])
y_pred_class = [1 if p < 1.5 else 2 for p in y_pred]
plt.scatter(X[:, 1], X[:, 2], c=y_pred_class)
```

```python
from sklearn.preprocessing import PolynomialFeatures

degree=2
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X[:, 1:3])
```

```python
model.fit(X_poly, y)
y_pred = model.predict(X_poly)
y_pred_class = [1 if p < 1.5 else 2 for p in y_pred]
plt.scatter(X[:, 1], X[:, 2], c=y_pred_class)
```

## サポートベクターマシン

```python
from sklearn.svm import SVC

model = SVC(kernel='rbf', C=1.0)  # C が大きいと正則化は弱まる、C が小さいと正則化が強く働く
model.fit(X[:, 1:3], y)
y_pred = model.predict(X[:, 1:3])
plt.scatter(X[:, 1], X[:, 2], c=y_pred)
```

```python
model = SVC(kernel='rbf', C=1000000)
model.fit(X[:, 1:3], y)
y_pred = model.predict(X[:, 1:3])
plt.scatter(X[:, 1], X[:, 2], c=y_pred)
```
